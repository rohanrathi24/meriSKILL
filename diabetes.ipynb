## Diabetes Patients

Analyzing data on diabetes patients means looking at their health, lifestyle, and medical history to find important information. This helps doctors create better treatment plans, spot risk factors, and improve how they take care of patients. By understanding things like age, income, and if patients take their medicine, doctors can give more personalized care and work on preventing diabetes problems in the future.

```python
#Target Analysis
#Diabetical and Non-diabetical Persons

df["Outcome"].value_counts()

#percentage distribution of the "Outcome" 
print(100 * df["Outcome"].value_counts() / len(df))

with_diabetes = df['Outcome'].value_counts()[1]
without_diabetes = df['Outcome'].value_counts()[0]
print(f"Patients with Diabetes: {with_diabetes}\nPatients without Diabetes: {without_diabetes}")

sns.countplot(x="Outcome", data=df)
plt.show()
```

```python
#visualizing Pregnancies avrage outcome
plt.figure(figsize=(8,6))
sns.countplot(x='Pregnancies', hue='Outcome', data = df)
plt.show()
```

```python
df.describe().T
```

```python
#visualizing the distribution of data in each column

plt.figure(figsize=(12, 6))
for i, col in enumerate(df.columns[:-1]):
    plt.subplot(2, 4, i + 1)
    sns.histplot(df[df['Outcome'] == 1][col], kde=True, label='Diabetes', color='blue')
    sns.histplot(df[df['Outcome'] == 0][col], kde=True, label='No Diabetes', color='orange')
    plt.title(f"Distribution of {col}")
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,8))

for i, col in enumerate(df.columns[:-1]):
    plt.subplot(2, 4, i + 1)
    sns.barplot(x='Outcome', y=df[col], data=df)
    plt.title(f"{col} vs. Outcome")

plt.tight_layout()
plt.show()
```

# Analysing Correlation

```python
df.corr()
```

```python
plt.figure(figsize=(10, 6))

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

Glucose vs. Outcome (0.467) -> There is a positive correlation between glucose levels and the outcome. This indicates that higher glucose levels are associated with a higher likelihood of having diabetes.

Age vs. Outcome (0.238) -> There is a weak positive correlation between age and our outcomes. This suggests that, on average, older individuals may have a slightly higher likelihood of having diabetes.

BMI vs. Glucose (0.221) -> There is a weak positive correlation between BMI and glucose levels. This suggests that individuals with higher BMI values tend to have slightly higher glucose levels.

Insulin vs. Glucose (0.331) -> There is a positive correlation between insulin levels and glucose levels.This indicates that higher glucose levels are associated with higher insulin levels.

BMI vs. SkinThickness (0.393) -> There is a positive correlation between BMI and skin thickness. This suggests that individuals with higher BMI values may have slightly thicker skin.

Age vs. Pregnancies (0.544) -> There is a positive correlation between age and the number of pregnancies. This confirms the earlier observation that older individuals tend to have more pregnancies.
 
SkinThickness vs. Insulin (0.437) -> There is a positive correlation between skin thickness and insulin levels. This suggests that individuals with thicker skin may have higher insulin levels.
